```python
# Carregar a biblioteca pandas

import pandas as pd

# Carregar a base de dados

aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')

# Visualizar tabela - Primeiras 10 linhas

aluguel_casa.head(10)
```

```python
# Visualizar tabela - Ultimas 10 linhas

aluguel_casa.tail(10)
```

```python
# Alterar nome das colunas

aluguel_casa = aluguel_casa.rename(
    columns={
      'index':'ID',
      'city':'Cidade',
      'area':'Area',
      'rooms':'Quartos',
      'bathroom':'Banheiros',
      'parking spaces':'Vagas_Estacionamento',
      'floor':'Piso',
      'animal':'Animais',
      'furniture':'Mobilia',
      'hoa':'HOA',
      'rent amount':'Valor_Aluguel',
      'property tax':'Taxa_Condominio',
      'fire insurance':'Seguro_Incendio',
      'total':'Total'
      }
    )

# Visualizar a tabela com as alterações nas colunas

aluguel_casa.head()
```

```python
#Visualizar tipo de dados

aluguel_casa.dtypes
```

```python
#Alterar tipo de dados

aluguel_casa = aluguel_casa.astype(
    {'Area':'float',
     'Valor_Aluguel':'float',
     'Taxa_Condominio':'float',
     'Seguro_Incendio':'float',
     'Total':'float'
    }
)

# Visualizar tipos de dados alterados

aluguel_casa.dtypes

# Visualizar tabela com tipo de dados atualizados

aluguel_casa.head()
```

```python
# Resumo dados

aluguel_casa.info()
```

```python
# Visualizar apenas algumas colunas

aluguel_casa[['Area',
              'Quartos',
              'Banheiros',
              'Vagas_Estacionamento',
              'Valor_Aluguel',
              'Taxa_Condominio',
              'Seguro_Incendio',
              'Total']]
```

```python
# Análise descritiva básica das colunas selecionadas

aluguel_casa[['Area',
              'Quartos',
              'Banheiros',
              'Vagas_Estacionamento',
              'Valor_Aluguel',
              'Taxa_Condominio',
              'Seguro_Incendio',
              'Total']].describe()
```

```python

# Filtrando dados

aluguel_casa = aluguel_casa[
    (aluguel_casa['Total']>=4000) & # Menor que 4000
    (aluguel_casa['Animais']=='accept') & # Aceita animais
    ((aluguel_casa['Quartos']==2)|(aluguel_casa['Quartos']==3)) # Com 2 ou 3 quartos
    ]

aluguel_casa.head()
```
