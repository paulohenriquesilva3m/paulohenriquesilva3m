```python
# Carregar a biblioteca pandas (analise de dados)

import pandas as pd

# Carregar a base de dados

aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')

# Visualizar tabela - Primeiras 10 linhas

aluguel_casa.head(10)
```

```python
# Visualizar tabela - Ultimas 10 linhas

aluguel_casa.tail(10)
```

```python
# Alterar nome das colunas

aluguel_casa = aluguel_casa.rename(
    columns={
      'index':'ID',
      'city':'Cidade',
      'area':'Area',
      'rooms':'Quartos',
      'bathroom':'Banheiros',
      'parking spaces':'Vagas_Estacionamento',
      'floor':'Piso',
      'animal':'Animais',
      'furniture':'Mobilia',
      'hoa':'HOA',
      'rent amount':'Valor_Aluguel',
      'property tax':'Taxa_Condominio',
      'fire insurance':'Seguro_Incendio',
      'total':'Total'
      }
    )

# Visualizar a tabela com as alterações nas colunas

aluguel_casa.head()
```

```python
#Visualizar tipo de dados

aluguel_casa.dtypes
```

```python
#Alterar tipo de dados

aluguel_casa = aluguel_casa.astype(
    {'Area':'float',
     'Valor_Aluguel':'float',
     'Taxa_Condominio':'float',
     'Seguro_Incendio':'float',
     'Total':'float'
    }
)

# Visualizar tipos de dados alterados

aluguel_casa.dtypes

# Visualizar tabela com tipo de dados atualizados

aluguel_casa.head()
```

```python
# Resumo dados

aluguel_casa.info()
```

```python
# Visualizar apenas algumas colunas

aluguel_casa[['Area',
              'Quartos',
              'Banheiros',
              'Vagas_Estacionamento',
              'Valor_Aluguel',
              'Taxa_Condominio',
              'Seguro_Incendio',
              'Total']]
```

```python
# Análise descritiva básica das colunas selecionadas

aluguel_casa[['Area',
              'Quartos',
              'Banheiros',
              'Vagas_Estacionamento',
              'Valor_Aluguel',
              'Taxa_Condominio',
              'Seguro_Incendio',
              'Total']].describe()
```

```python

# Filtrando dados

aluguel_casa = aluguel_casa[
    (aluguel_casa['Total']<=4000) & # Menor que 4000
    (aluguel_casa['Animais']=='accept') & # Aceita animais
    ((aluguel_casa['Quartos']==2)|(aluguel_casa['Quartos']==3)) # Com 2 ou 3 quartos
    ]

aluguel_casa.tail()
```

```python
# Nova análise descritiva básica das colunas selecionadas

aluguel_casa[['Area',
              'Quartos',
              'Banheiros',
              'Vagas_Estacionamento',
              'Valor_Aluguel',
              'Taxa_Condominio',
              'Seguro_Incendio',
              'Total']].describe()
```

```python
# Filtrando dados com a função .query

aluguel_casa = aluguel_casa.query(('Vagas_Estacionamento >=2 and Mobilia == "furnished"'))

aluguel_casa.head()

```

```python
# Nova análise descritiva básica das colunas selecionadas

aluguel_casa[['Area',
              'Quartos',
              'Banheiros',
              'Vagas_Estacionamento',
              'Valor_Aluguel',
              'Taxa_Condominio',
              'Seguro_Incendio',
              'Total']].describe()
```

```python
# Agrupando dados

# Agrupado por Cidade e preço médio do aluguel (Total)

# mean = média      #reset_index = formatação

agg_Cidade = aluguel_casa.groupby(['Cidade'])['Total'].agg('mean')
agg_Cidade.reset_index()
```

```python
# Agrupando dados

# Agrupado por quantidade de Quartos e Banheiros e preço médio do aluguel (Total)

# mean = média      #reset_index = formatação

agg_Quartos = aluguel_casa.groupby(['Quartos','Banheiros'])['Total'].agg('mean').reset_index()

agg_Quartos.reset_index()
```

```python
# Visualização dos dados através de gráficos

import seaborn as sns # biblioteca de visualização

ax = sns.barplot(data=agg_Quartos, x='Quartos', y='Total', color='black') #gráfico de barras

ax.set(xlabel = "Quantidade de quartos",
       ylabel = 'Média de preço');


```

```python
# Histograma

agg_Area = aluguel_casa[['Area']]


ay = sns.histplot(agg_Area, kde=True); #kde = linha de tendência
ay.set(xlabel = 'Área (m²)',
       ylabel = 'Frequência');
```

```python
# Carregar a base de dados

aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')

# Alterar nome das colunas

aluguel_casa = aluguel_casa.rename(
    columns={
      'index':'ID',
      'city':'Cidade',
      'area':'Area',
      'rooms':'Quartos',
      'bathroom':'Banheiros',
      'parking spaces':'Vagas_Estacionamento',
      'floor':'Piso',
      'animal':'Animais',
      'furniture':'Mobilia',
      'hoa':'HOA',
      'rent amount':'Valor_Aluguel',
      'property tax':'Taxa_Condominio',
      'fire insurance':'Seguro_Incendio',
      'total':'Total'
      }
    )

#Alterar tipo de dados

aluguel_casa = aluguel_casa.astype(
    {'Area':'float',
     'Valor_Aluguel':'float',
     'Taxa_Condominio':'float',
     'Seguro_Incendio':'float',
     'Total':'float'
    }
)

# Visualizar tipos de dados alterados

aluguel_casa.dtypes

# Visualizar tabela com tipo de dados atualizados

aluguel_casa.head()
```

```python
# Histograma

agg_Area = aluguel_casa[['Area']]

agg_Area = agg_Area.query(('Area <=100'))

ay = sns.histplot(agg_Area, kde=True); #kde = linha de tendência
ay.set(xlabel = 'Área (m²)',
       ylabel = 'Frequência');
```

```python
# Histograma

agg_Total = aluguel_casa[['Total','Cidade']]

agg_Total = agg_Total.query(('Total <= 4000'))

ay = sns.histplot(data = agg_Total, x='Total', hue = 'Cidade', kde=True); #kde = linha de tendência
ay.set(xlabel = 'Valor',
       ylabel = 'Frequência');
```

```python
# Histograma

agg_Piso = aluguel_casa[['Total','Cidade','Piso']]

agg_Piso = agg_Piso.query(('Total <= 4000 and Cidade == "Porto"'))

agg_Piso = agg_Piso.groupby('Piso')['Total'].agg('mean').reset_index()

sns.barplot(data = agg_Piso, x='Piso', y='Total');





```
